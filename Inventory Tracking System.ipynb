```python
dt2 = {
    "Item_ID": [101, 102, 103],
    "Item_Name": ["Laptop", "Mouse", "Keyboard"],
    "Category": ["Electronics", "Electronics", "Electronics"],
    "Quantity": [25, 150, 120],
    "Unit_Price": [55000, 450, 850]
}
dt2
```

```python
while True:
    print("\n----- Inventory Management Menu -----")
    print("1. View Inventory")
    print("2. Add Item")
    print("3. Search Item")
    print("4. Delete Item")
    print("5. Exit")

    choice = input("Enter your choice: ")

    # View Inventory
    if choice == "1":
        print("\nInventory Details:")
        for i in range(len(dt2["Item_ID"])):
            print(
                dt2["Item_ID"][i],
                dt2["Item_Name"][i],
                dt2["Category"][i],
                dt2["Quantity"][i],
                dt2["Unit_Price"][i]
            )

    # Add Item
    elif choice == "2":
        dt2["Item_ID"].append(int(input("Enter Item ID: ")))
        dt2["Item_Name"].append(input("Enter Item Name: "))
        dt2["Category"].append(input("Enter Category: "))
        dt2["Quantity"].append(int(input("Enter Quantity: ")))
        dt2["Unit_Price"].append(int(input("Enter Unit Price: ")))
        print("Item added successfully!")

    # Search Item
    elif choice == "3":
        search_id = int(input("Enter Item ID to search: "))
        if search_id in dt2["Item_ID"]:
            index = dt2["Item_ID"].index(search_id)
            print("Item Found:")
            print(
                dt2["Item_ID"][index],
                dt2["Item_Name"][index],
                dt2["Category"][index],
                dt2["Quantity"][index],
                dt2["Unit_Price"][index] )
        else:
            print("Item not found!")

    # Delete Item
    elif choice == "4":
        del_id = int(input("Enter Item ID to delete: "))
        if del_id in dt2["Item_ID"]:
            index = dt2["Item_ID"].index(del_id)
            for key in dt2:
                dt2[key].pop(index)
            print("Item deleted successfully!")
        else:
            print("Item not found!")

    # Exit
    elif choice == "5":
        print("Exiting Inventory System...")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
def view_inventory():
    print("\nInventory Details:")
    for i in range(len(dt2["Item_ID"])):
        print(
            dt2["Item_ID"][i],
            dt2["Item_Name"][i],
            dt2["Category"][i],
            dt2["Quantity"][i],
            dt2["Unit_Price"][i] )
def add_item():
    dt2["Item_ID"].append(int(input("Enter Item ID: ")))
    dt2["Item_Name"].append(input("Enter Item Name: "))
    dt2["Category"].append(input("Enter Category: "))
    dt2["Quantity"].append(int(input("Enter Quantity: ")))
    dt2["Unit_Price"].append(int(input("Enter Unit Price: ")))
    print("Item added successfully!")
def search_item():
    search_id = int(input("Enter Item ID to search: "))
    if search_id in dt2["Item_ID"]:
        index = dt2["Item_ID"].index(search_id)
        print("Item Found:")
        print(
            dt2["Item_ID"][index],
            dt2["Item_Name"][index],
            dt2["Category"][index],
            dt2["Quantity"][index],
            dt2["Unit_Price"][index]
        )
    else:
        print("Item not found!")

# -------- Menu-driven Program --------
while True:
    print("\n--- Inventory Menu ---")
    print("1. View Inventory")
    print("2. Add Item")
    print("3. Search Item")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        view_inventory()
    elif choice == "2":
        add_item()
    elif choice == "3":
        search_item()
    elif choice == "4":
        break
    else:
        print("Invalid choice!")
```

```python
# Inventory Dictionary
dt2 = {
    "Item_ID": [101, 102, 103],
    "Item_Name": ["Laptop", "Mouse", "Keyboard"],
    "Category": ["Electronics", "Electronics", "Electronics"],
    "Quantity": [25, 150, 120],
    "Unit_Price": [55000, 450, 850]
}

# -------- Function to Add Item --------
def add_item():
    name = input("Enter Item Name: ").strip().title()   # string methods

    dt2["Item_ID"].append(104)          # list method
    dt2["Item_Name"].append(name)
    dt2["Category"].append("Electronics")
    dt2["Quantity"].append(50)
    dt2["Unit_Price"].append(1000)

    print("Item added successfully!")

# -------- Function to Delete Item --------
def delete_item():
    if len(dt2["Item_ID"]) > 0:          # list method
        index = 0
        for key in dt2.keys():           # dictionary method
            dt2[key].pop(index)          # list method
        print("Item deleted.")

# -------- Function to Display Inventory --------
def view_inventory():
    print("\nInventory Keys:", dt2.keys())      # dictionary method
    print("Inventory Values:", dt2.values())    # dictionary method

    for i in range(len(dt2["Item_ID"])):
        print(
            dt2.get("Item_Name")[i],             # dictionary method
            dt2["Quantity"][i]
        )

# -------- Main Program --------
add_item()
view_inventory()
delete_item()
```
